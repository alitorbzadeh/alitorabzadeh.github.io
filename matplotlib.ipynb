# **Matplotlib package**

```python
import matplotlib.pyplot as plot
import numpy as np
```

# رسم یک نمودار دوبعدی ساده

```python
x=np.arange(1,10,1)
y=np.arange(2,11,1)
x1=np.arange(1,10,2)
y1=np.arange(1,10,2)
plot.plot(x,y,color="green",marker="d",linestyle="dashed")
plot.plot(x1,y1,color="red",marker="s",linestyle="dashed")
plot.show
plot.xlabel("x ha")
plot.ylabel("y ha")
plot.title("a sample for start")
plot.grid()
plot.legend("blue")
plot.xlim(2,8)
plot.ylim(2,8)
```

# رسم نمودار تابع در پایتون

```python
x=[0,(np.pi)/4,(np.pi)/2,(3*(np.pi))/4,(np.pi),(5*(np.pi))/4,(3*(np.pi))/2,2*(np.pi)]
y=np.cos(x)
plot.plot(x,y)
plot.grid()
```

```python
x1=np.arange(-10,10,.001)



def function(inp):
  x1=np.arange(-10,10,.00001)
  if inp=="leaner":
    y0=x1
  elif inp=="daraje2":
    y0=x1**2
  elif inp=="daraje3":
    y0=x1**3
  elif inp=="daraje4":
    y0=x1**4
  x=x1
  y=y0
  return(x,y)

fig=plot.figure()
plt1=fig.add_subplot(221)
plt2=fig.add_subplot(222)
plt3=fig.add_subplot(223)
plt4=fig.add_subplot(224)

(x,y)=function("leaner")
plt1.plot(x,y,color="red")
plt1.set_title('$y_1 = x$')
plt1.grid()

(x,y)=function("daraje2")
plt2.plot(x,y,color="blue")
plt2.set_title('$y_2 = x^2$')
plt2.grid()

(x,y)=function("daraje3")
plt3.plot(x,y,color="green")
plt3.set_title('$y_3 = x^3$')
plt3.grid()

(x,y)=function("daraje4")
plt4.plot(x,y,color="black")
plt4.set_title('$y_4 = x^4$')
plt4.grid()

fig.subplots_adjust(hspace=.5,wspace=0.5)
```
